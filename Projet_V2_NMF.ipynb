```python
'''
python class for the data engineering (reading, cleaning, training...)
localised in ./Python
'''
from tagText import TagText 
```

```python
tagText = TagText()
print(tagText.urlDirectory)#folder for cvs file generated by SQL query (cf. I) 
print(tagText.fileName)#name of the cvs file generated by SQL query (cf. I)
data_question = tagText.read_source()
data_question.head()

tagText.nombre_post_entree = 500
tagText.precision = 500
```

```python
data_preprocessed = data_question['body'].apply(tagText.preprocessing)
print(data_preprocessed[1:9])
```

```python
tableau_perf = tagText.nmf_find_topic_number(data_preprocessed, 5, 200, 5)
print(tableau_perf)
```

```python
import matplotlib.pyplot as plt
import numpy as np
mini=min(tableau_perf)
#print(perfs)
#print(perfs-mini)
x = range(5, 200, 5)

from sklearn.preprocessing import scale
perfs_normalized = scale( tableau_perf, axis=0, with_mean=True, with_std=True, copy=True )
x_normalized = scale( x, axis=0, with_mean=True, with_std=True, copy=True )


plt.bar(x, tableau_perf)
plt.ylabel("Performance")
plt.xlabel("number of topics")
plt.title("performance/number of topics")
plt.show()

plt.bar(x, perfs_normalized)
plt.ylabel("Performance normalized")
plt.xlabel("number of topics")
plt.title("performance/number of topics")
plt.show()

perfs_topic_optimized = np.multiply(x_normalized, perfs_normalized)
plt.bar(x, perfs_topic_optimized)
plt.ylabel("Performance normalized * number of topic * coef")
plt.xlabel("number of topics")
plt.title("performance/number of topics")
plt.show()


perfs_topic_optimized = np.subtract(x_normalized/2, perfs_normalized)
plt.bar(x, perfs_topic_optimized)
plt.ylabel("number of topic optimized - Performance normalized")
plt.xlabel("number of topics")
plt.title("performance/number of topics")
plt.show()


perfs_topic_optimized = np.add(x_normalized, perfs_normalized)
plt.bar(x, perfs_topic_optimized)
plt.ylabel("number of topic optimized + Performance normalized")
plt.xlabel("number of topics")
plt.title("performance/number of topics")
plt.show()

plt.bar(x, tableau_perf-mini)
plt.ylabel("Delta Performance")
plt.xlabel("number of topics")
plt.title("performance/number of topics")
plt.show()
'''
plt.bar(x, np.subtract(perfs,x))
plt.ylabel("Performance - number of topic")
plt.xlabel("number of topics")
plt.title("performance/number of topics")
plt.show()
'''


```
